```python
# Task 1 — Data Parsing & Profile Cleaning 
# Given dataset raw_students contains dictionaries within a list.
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
# A new list is created with all the inconsistencies removed
cleaned_students = []

for student in raw_students:
    
    clean_one = student["name"].strip() # by using .strip(), all the extra unwanted spaces at the beginning and end of a string are removed
    clean_peeps = clean_one.title() # by using .title(), the names(strings) are converted into title case
    print(clean_peeps)

    clean_roll = int(student["roll"]) # here, the roll numbers, that are stored as strings, are converted into integers
    print(clean_roll)

    marks_one = student["marks_str"].split(", ") # here, each marks are split by a comma and stored individually
    
    clean_marks = [] # a new empty list is made to store cleaned marks
    
    for m in marks_one:
        marks = int(m)            # here, all the individual marks stored as string, are converted into integers
        clean_marks.append(marks)     #and these integer marks are stored in the list clean_marks that we made earlier
    print(clean_marks)

    name_check = clean_peeps.replace(" ", "") # to check whether a name is valid or not, we first clean spaces between first and last name

    if name_check.isalpha(): #.isalpha() is used to check whether the name is alphabetical or not
        print("✓ Valid name\n") # it is a valid name if the name contains only alphabets
    else:
        print("✗ Invalid name\n") # it is not a valid name if the name contains other characters like numbers or special characters or spaces.
                                    # this is why, the spaces between first and last name was removed, or it would show the name as invalid
    new_profile = {
        "name": clean_peeps,
        "roll": clean_roll,
        "marks": clean_marks
    }
    cleaned_students.append(new_profile)
    # a new dictionary is created that stores all the data that is now cleaned
print(cleaned_students)
```

```python
# to print profile cards. using f strings makes it easier, we do not have to make different profile cards for every student. 
for student in cleaned_students:
    print("================================")
    print(f"Student : {student['name']}")
    print(f"Roll No : {student['roll']}")
    print(f"Marks   : {student['marks']}")
    print("================================\n")
```

```python
# a for loop is used to search through the list of dictionaries of cleaned students, to get the information of a certain student with roll number 103.
for s in cleaned_students:
    if s["roll"] == 103:
        #target_name = s["name"]
        print(f"Roll 103 Uppercase: {s["name"].upper()}") # .upper() is used to convert string into UPPER case
        print(f"Roll 103 Lowercase: {s["name"].lower()}") # .lower() is used to convert srting into lower case
```

```python
# Task 2 — Marks Analysis Using Loops & Conditionals
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# enumerate is used, to give 2 things at once. in this case, enumerating subjects can allow us also look into another list (marks) at the same time.
for i, sub in enumerate(subjects): 
    score = marks[i]
    
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"
        
    print(f"{sub}: {grade}")

# here, we use conditionals like if, elif and else, to grade Ayesha Sharma's marks into categories, A+, A, B, C and F
# so we tell the computer that, look at Ayesha's marks and compare them to this scheme: if marks are greater than or equal to 90 give her an A+.
# or if marks are between 80 and 90, give her an A grade. and so on...
```

```python
total = sum(marks) # we use sum() to add up the marks and get a total.
avg = total / len(marks) # we use len() to count the number of items/entries in the list.

high_score = max(marks)
high_index = marks.index(high_score)
high_sub = subjects[high_index]

low_score = min(marks)
low_index = marks.index(low_score)
low_sub = subjects[low_index]

# to get the highest scoring subject, we first find the highest marks using the max() function, that gets us the maximum value.
# then we are saying, get the index(position) of this highest number and find the subject name sitting in that index/position and get it's name.
# Similarly for finding the lowest scoring subject, we use the min() function to find the minimum value from the marks,
# get it's index and using that index find the subject name.

print(f"\nTotal Marks: {total}")
print(f"Average: {avg:.2f}") # Using .2f for 2 decimal places as per notes
print(f"The subject with the highest score is: {high_sub}, with a score of {high_score}")
print(f"The subject with the lowest score is: {low_sub}, with a score of {low_score}")
```

```python
new_count = 0 # we are initializing a count variable to 0

while True: # while True is used to start a continuous loop.
    input_one = input("\nEnter subject name (or type 'done' to finish): ") # we ask the user to enter the name of a subject
    input_two = input_one.strip() # we are trying to clean the data entered by the user spontaneously to remove any extra spaces using .strip() function
    sub_input = input_two.title() # we are trying to convert the input given into title case

    if sub_input.lower() == "done": # here, if the user types in 'done', we break the loop and end
        break
    
    if not sub_input.replace(" ", "").isalpha(): #if the subject name entered by the user is not in alphabets, we show an error.
        print("Invalid subject name. Please enter in alphabetical letters only.")
        continue

    while True:
        mark_input = input(f"Enter marks for {sub_input}: ") # next, if the subject name is valid, we ask the user to enter the subject marks
        
        is_valid_num = mark_input.isdigit() or (mark_input.startswith('-') and mark_input[1:].isdigit())
        
        if not is_valid_num:
            print("I get what you are saying, but I don't get what you are saying. Enter marks in integers please.")
            continue 

        val = int(mark_input) # if the marks entered is numeric, we check whether they are between the range of 0 to 100.

        if val < 0:
            print("Quit joking, we know it's not that bad. Enter marks between 0 and 100 only.")
        elif val > 100:
            print("Whoa! You are either a crazy genius or made a mistake. Enter marks between 0 and 100 only.")
        else:
            subjects.append(sub_input) #if the all the entries are valid, we save these inputs and increase the count we initialized earlier
            marks.append(val)
            new_count += 1
            print(f"{sub_input} added successfully.")
            break 
    
final_avg = sum(marks) / len(marks) #we calculate the new average after adding the new sujects and marks
print(f"\nNumber of new subjects added: {new_count}")
print(f"New Overall Average: {final_avg:.2f}")
```

```python
# Task 3 — Class Performance Summary
class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma",   [55, 68, 49, 72, 61]),
    ("Priya Nair",    [91, 85, 88, 94, 79]),
    ("Karan Mehta",   [40, 55, 38, 62, 50]),
    ("Sneha Pillai",  [75, 80, 70, 68, 85]),
]

pass_count = 0 # we initialize variables to count students passed and students failed
fail_count = 0
all_averages = [] # we initialize a new list to store the average marks

for name, marks in class_data:
    
    avg = sum(marks) / len(marks) # to compute average marks and then save that average in the list
    all_averages.append(avg)
    
    # to determine whether a student has passed or failed and update the counters
    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else: 
        status = "Fail"
        fail_count += 1

print(f"Number of students passed: {pass_count}")
print(f"Number of students failed: {fail_count}")
print(f"Average scores of students: {all_averages}")
```

```python
# to print a class report
topper_name = ""
topper_avg = -1 

#to print table headings
print(f"{'Name':<18} | {'Average':<7} | {'Status'}")
print("-" * 40)

# we loop through the data and first calculate average marks per student. if the calculated average is above 60, we declare the sudent as pass or else fail.
# to find the topper average, through the loop, we compare the scores, and update the topper_avg counter we set before.
# by the end of the loop, the topper_avg will hold the maximum value. and that topper_name will hold the name of the student with the top average marks
for name, marks in class_data:   
    avg = sum(marks) / len(marks)
    if avg >= 60:
        status = "Pass"
    else:
        status = "Fail"
    
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name
        
    print(f"{name:<18} | {avg:>7.2f} | {status}") # we enter the details of each student in this loop, in the table
```

```python
class_avg = sum(all_averages) / len(all_averages)

print(f"Passed: {pass_count} | Failed: {fail_count}")
print(f"Class Topper: {topper_name} ({topper_avg:.2f})")
print(f"Overall Class Average: {class_avg:.2f}")
```

```python
# Task 4 — String Manipulation Utility
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# 1. to remove blankspace using .strip()
clean_essay = essay.strip()
print(f"Cleaned Essay: {clean_essay}\n")

# 2. to convert the essay into title case using .title()
print(f"Title Case Essay: {clean_essay.title()}\n")

# 3. to count the number of times the word python is used
print(f"Python Count: {clean_essay.count("python")}\n")

# 4. to replace "python" with "Python 🐍" in the clean_essay
print(f"Essay with Replaced python: {clean_essay.replace("python", "Python 🐍")}\n")

# 5. to split into sentences we split on ". " to get a list of strings
sentence_list = clean_essay.split(". ")
print(f"Sentence List: {sentence_list}\n")

# 6. to get numbered sentences and print each sentence on a new line
for i, sentence in enumerate(sentence_list, 1):
    s = sentence.strip() # we remove any blankspaces at the start or end os the sentences
    if not s.endswith("."): # then we check whether or not the sentence ends with a . , if not, we add it
        s += "."
        
    print(f"{i}. {s}")
```
